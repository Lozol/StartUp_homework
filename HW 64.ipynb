```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
im_src = cv2.imread('chess_1.png')
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_src)
plt.show()
```

```python
pts_src = np.array([[457, 41], [458, 261], [193, 50],[196, 246]])
```

```python
im_dst = cv2.imread('chess_2.png')
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_src)
plt.show()
```

```python
pts_dst = np.array([[164, 35],[530, 103],[188, 318],[407, 365]])
```

```python
h, status = cv2.findHomography(pts_src, pts_dst)
h
```

```python
im_out = cv2.warpPerspective(im_src, h,
                            (im_dst.shape[1],
                            im_dst.shape[0]))
```

```python
plt.figure(figsize=(12,8))
plt.imshow(im_out)
plt.show()
```
