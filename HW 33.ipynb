## Task 2

Візуалізуйте залежності: споживання палива від вартості - лінійною діаграмою; споживання палива від швидкості - точковим графіком; вантажопідйомність від швидкості - точковим графіком.

```python

a = df['petrol']
b = df['cost']
c = df['max_sp']
d = df['load']

fig = plt.figure(figsize =(10, 12))


# Две строки, два столбца. Текущая ячейка - 1
plt.subplot (3, 1, 1)
plt.plot(a,b, alpha=0.8, color='blue')
plt.title ("споживання палива від вартості")
plt.xlabel('паливо') 
plt.ylabel('вартість')

# Две строки, два столбца. Текущая ячейка - 3
plt.subplot (3, 1, 2)
plt.scatter(a, c, alpha=0.8,color='blue')    #plot (xlist, ylist)
plt.title ("споживання палива від швидкості")
plt.xlabel('паливо') 
plt.ylabel('макс швидкість')


# Две строки, два столбца. Текущая ячейка - 2
plt.subplot (3, 1, 3)
plt.scatter(d, c, alpha=0.8,color='blue')
plt.title (" вантажопідйомність від швидкості")
plt.xlabel('вантажопідйомність') 
plt.ylabel('макс швидкість')

plt.show()
```
