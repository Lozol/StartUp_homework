```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
```

```python
img = tf.keras.utils.load_img('car1.jpg')
```

```python
img2 = image_loaded_2 = tf.keras.utils.load_img('car1.jpg',
                                         target_size=(200,
                                                      300))
```

```python
print("Size of Image 1: ", img.size)
print("Size of Image 2: ", img2.size)
```

```python
label_name = 'norm_car'
plt.imshow(img2)
plt.title(label_name)
```

```python
def visualize(original, augmented):
    fig = plt.figure()
    plt.subplot(1,2,1)
    plt.title('Original image')
    plt.imshow(original)
    plt.axis("off")
 
    plt.subplot(1,2,2)
    plt.title('Augmented image')
    plt.imshow(augmented)
    plt.axis("off")
```

```python
flipped = tf.image.flip_left_right(img)
visualize(img, flipped)
```

```python
saturation = tf.image.adjust_saturation(img, 3)
visualize(img, saturation)
```

```python
gray = tf.image.rgb_to_grayscale(img)
visualize(img, gray)
```

```python
rotated = tf.image.rot90(img)
visualize(img, rotated)
```

```python
 cropp = tf.image.central_crop(img, central_fraction=0.9)
visualize(img, cropp)
```

```python
brightness = tf.image.adjust_brightness(img, 0.5)
visualize(img, brightness)
```
