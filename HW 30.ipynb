На рівні значущості 0,01 перевіремо гіпотезу про нормальний розподіл ознаки  генеральної сукупності проти конкурентної гіпотези  про те що вона так НЕ розподілена. Використовуючи критерій узгодження Пірсона.

для  8ступнів свободи табл значення 20,1, того мі відхиляемо нульву гіпотезу

```python

data = ([12,36,36,60,60,60,60,84,84,84,84,84,84,108,108,108,108,108,108,108,108,108,108,108,108,
       132,132,132,132,132,132,132,132,132,132,132,132,132,132,132,132,
       156,156,156,156,156,156,180,180,204])

stat, p = sps.normaltest(data)  # Критерій узгодження Пірсона
print('Statistics=%.3f, p-value=%.3f' % (stat, p))

alpha = 0.01
if p < alpha:
    print('Приймаємо гіпотезу про нормальність розподілу')
else:
    print('Не приймаємо гіпотезу про нормальність розподілу')
```
